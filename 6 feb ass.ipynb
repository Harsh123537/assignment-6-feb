```python
def fun(l):
    lis=[]
    for i in l:
        if type(i)==bool:
            continue
        elif type(i)==int:
            lis.append(i)
        else:
            if type(i)==dict:
                for j,k in i.items():
                    if type(j)==int:
                        lis.append(j)
                    if type(k)==int:
                        lis.append(k)
                    else:
                        if type(k)==bool:
                            continue
                        else:
                            for m in k:
                                if type(m)==int:
                                    lis.append(m)
            else:
                for j in i:
                    if type(j)==int:
                        lis.append(j)
    from functools import reduce
    return reduce(lambda x,y:x*y,lis),lis
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,  22, 61, 34)}, [56, 'data science'], 'Machine Learning'] 
print(fun(list1))
    
```

```python
s="I want to become a Data Scientist."
s.lower()
s=s.split(" ")
l=[]
for i in s:
    for j in i:
        m=j.split()
        for p in m:
            l.append(p)
    l.append(" ")

for i in range(len(l)-1):
    if l[i]=='a':
        l[i]='z'
    elif l[i]=='b':
        l[i]='y'
    elif l[i]=='c':
        l[i]='x'
    elif l[i]==' ':
        l[i]='$'
s=''.join(l)
print(s)
```
